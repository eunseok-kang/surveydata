```python
import pandas as pd
import numpy as np
import math
import matplotlib
urlfile="https://raw.githubusercontent.com/fivethirtyeight/data/master/congress-age/congress-terms.csv"
df=(pd.read_csv(urlfile))
print(df)
print(df.head())
```

```python
congress_mean = df.groupby('party')[['congress']].mean()
congress_median = df.groupby('party')[['congress']].median()
print(congress_mean)
print(congress_median)
```

```python
age_mean  = df.groupby('party')[['age']].mean()
age_median = df.groupby('party')[['age']].median()
print(age_mean)
print(age_median)
```

```python

x=df['age']
y=df['congress']
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
df.plot(kind = 'scatter', title='Age relation to # Congress in History', ylabel = '# Congress', xlabel = 'Age', figsize=(20, 5), y='congress', x='age')
```
